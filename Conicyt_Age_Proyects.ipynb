```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
#Import the data, using Pandas
projects_age = pd.read_csv("ResumenEdad.txt", sep="\t",index_col="Codigo")
```

```python
#Check that everything looks ok
projects_age
```

```python
#Mean by proposal type
projects_age.groupby("tipo_concurso")["edad"].mean()
```

```python
#Median by proposal type
projects_age.groupby("tipo_concurso")["edad"].median()
```

```python
#Summary by proposal type
projects_age.groupby("tipo_concurso")["edad"].describe()
```

```python
#summary years 2006-2013
rango_2006_2013 = [2006,2007,2008,2009,2010,2011,2012,2013]
projects_age[projects_age.year.isin(rango_2006_2013)].groupby("tipo_concurso")["edad"].describe()
```

```python
#summary years 2006-2013
projects_age[projects_age.year.isin(rango_2006_2013)].groupby("tipo_concurso")["edad"].max()
```

```python
#Histogramas por tipo de proyecto. Cada linea es un año

tipo_concurso = ["INICIACION","POSTDOCTORADO","REGULAR"]

fig,axes = plt.subplots(3,1,figsize=(20,10))

for index, concurso in enumerate(tipo_concurso):
    edades = projects_age[projects_age.tipo_concurso == concurso].edad
    edad_max = edades.max()
    edad_min = edades.min()
    rango = edad_max-edad_min
        
    axes[index].hist(edades, rango, color=sns.desaturate("skyblue", .75))
    
    axes[index].axvline(x=edades.mean(),linewidth=3, color='mediumpurple')
        
    axes[index].set_xlim(25,85)
    axes[index].set_title(concurso,fontsize=18)
    axes[index].set_xlabel("Edad",fontsize=14)
    axes[index].set_ylabel("Numero de proyectos",fontsize=14)
    axes[index].tick_params(axis="both", labelsize=17)

fig.tight_layout()
fig.savefig("/Users/juan/Documents/Websites/octopress/source/images/blog_posts/011814/HistogramaConcursos_1982_2013.png")
```

```python
#Histogramas por tipo de proyecto. Cada linea es un año. Periodo 2006-2013

tipo_concurso = ["INICIACION","POSTDOCTORADO","REGULAR"]

fig,axes = plt.subplots(3,1,figsize=(20,10))

projects_age[projects_age.year.isin(rango_2006_2013)].groupby("tipo_concurso")["edad"].median()

for index, concurso in enumerate(tipo_concurso):
    edades = projects_age[(projects_age.tipo_concurso == concurso) & (projects_age.year.isin(rango_2006_2013))].edad
    edad_max = edades.max()
    edad_min = edades.min()
    rango = edad_max-edad_min
        
    axes[index].hist(edades, rango, color=sns.desaturate("skyblue", .75))
    
    axes[index].axvline(x=edades.mean(),linewidth=3, color='mediumpurple')
        
    axes[index].set_xlim(25,70)
    axes[index].set_title(concurso,fontsize=18)
    axes[index].set_xlabel("Edad",fontsize=14)
    axes[index].set_ylabel("Numero de proyectos",fontsize=14)
    axes[index].tick_params(axis="both", labelsize=17)

fig.tight_layout()
fig.savefig("/Users/juan/Documents/Websites/octopress/source/images/blog_posts/011814/HistogramaConcursos_2006_2013.png")
```

```python
#Comparacion de los diferentes concursos, usando boxplots y marcando cada punto (para tener una idea de las
#distribuciones
#Este grafico no fue usado en el post. Es solo para probar algunas cosas
fig, axes = plt.subplots(figsize=(15,10))
projects_age.boxplot(ax=axes, column="edad",by="tipo_concurso",grid=True)

j = 1
for i in ["INICIACION","POSTDOCTORADO","REGULAR"]:
    y = projects_age.edad[projects_age.tipo_concurso==i].dropna()
    
    #Add some random jitter to x-axis
    x = np.random.normal(j, 0.04, size=len(y))
    plt.plot(x,y, 'r.', alpha=0.2)
    j += 1

axes.set_xlabel("Edad", fontsize=18)
axes.set_ylabel("Concurso", fontsize=18)
axes.tick_params(axis="both", labelsize=15)
```

```python
#Comparacion de las distribuciones, usando violinplots en vez de boxplots
fig, axes = plt.subplots(figsize=(15,10))
#axes.title("Distribución de edades, concursos Conicyt")
axes.set_title("Distribucion de edades, Concursos Conicyt (1982-2013)",fontsize=25)
axes.tick_params(axis="both", labelsize=15)
sns.violinplot(projects_age.edad,projects_age.tipo_concurso, color="Paired", ax=axes,bw=1)
sns.axlabel("Concurso","Edad",fontsize=20)
fig.savefig("ConcursosCompDist_1982-2013.pdf")
```

```python
#Comparacion de las distribuciones, usando violinplots en vez de boxplots. Concurso de Iniciacion
fig, axes = plt.subplots(figsize=(18,10))

datos_iniciacion = projects_age[projects_age["tipo_concurso"] == "INICIACION"]

number_years = len(datos_iniciacion.groupby("year"))

axes.set_title(u"Proyectos de Iniciación",fontsize=23)
plt.xticks(rotation=45)
axes.tick_params(axis="both", labelsize=18)
sns.violinplot(datos_iniciacion.edad, datos_iniciacion.year,positions=range(1,number_years + 1,1),bw=0.5, ax=axes,color="Set3")

axes.plot(range(1,number_years + 1,1),datos_iniciacion.groupby("year")["edad"].mean(),linewidth=3)

#axes.xticks(range(1,number_years + 1,1), 52, datos_iniciacion.groupby("year"),ha="center", va="bottom")


#Add number of projects per year
i = 1
for edad_max, numero_projects in zip(datos_iniciacion.groupby("year")["edad"].max(), 
                                     datos_iniciacion.groupby("year")["edad"].count()):
    axes.text(i,edad_max + 2, numero_projects, fontsize=20, ha="center",va="center")
    i += 1


sns.axlabel(u"Año", "Edad", fontsize=20)

plt.ylim([25,55])

fig.tight_layout()
fig.savefig("/Users/juan/Documents/Websites/octopress/source/images/blog_posts/011814/DistIniciacion.png")

```
